```python
# imports
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
# make visualizations easier - jupyter only
```

```python
# creating some random data and ranges to plot 
```

```python
x_values = range(50)
```

```python
y_values = np.random.randint(0,100,50)
```

```python
plt.plot(x_values, y_values)
```

```python
plt.plot(x_values, y_values, "y--") # yellow dashed
plt.xlabel("range")
plt.ylabel("random values")
plt.title("This Is A Title")
```

```python
def label_and_title(): # just to make it a little bit easier
    plt.xlabel("range")
    plt.ylabel("random values")
    plt.title("This Is A Title")
```

```python
y_values_again = np.random.randint(30,60,50)
plt.plot(x_values, y_values, "y--") # yellow dashed
plt.plot(x_values, y_values_again, "g") # green dotted
label_and_title()
```

```python
y_values_again = np.random.randint(30,60,50)
plt.plot(x_values, y_values, "y--", label="0-100") # yellow dashed
plt.plot(x_values, y_values_again, "g", label="30,60") # green dotted
plt.legend(loc="upper left") # location for the labels
'''
    best
	upper right
	upper left
	lower left
	lower right
	right
	center left
	center right
	lower center
	upper center
	center
    '''
label_and_title()
```

```python
linspaceArray = np.linspace(0,100,50)
plt.plot(x_values, y_values, "y--") # yellow dashed
plt.plot(x_values, y_values_again, "g") # green line default
plt.plot(x_values, linspaceArray,"o", alpha=0.5) # default color dotted , alpha value for transparency
label_and_title()
plt.show()
```

### Histogram Graphs

```python
hist_values = np.random.randint(0,100,1000)
plt.hist(hist_values, bins=100)
plt.show()
```

### Scatter Plot

```python
plt.scatter(x=range(20), y=np.linspace(0,50,20) ** 2, c="orange")
```

### BoxPlot

```python
plt.boxplot(y_values) # seaborn prefered
plt.show()
```

### SUBPLOTS

```python
# plt.subplot(row,columns, plot_number)
plt.subplot(1,2,1)
plt.subplot(1,2,2)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x_values, y_values, "y--") # yellow dashed
plt.subplot(1,2,2)
plt.plot(x_values, y_values, "r--") # red dashed
plt.show()
```

### Figure Specified

```python
fig = plt.figure()
axes = fig.add_axes([0.0,0.0,0.8,0.8])
axes.plot(x_values, y_values_again, "g")
```

graph in graph

```python
fig = plt.figure()
axes1 = fig.add_axes([0.0,0.0,0.8,0.8])
axes1.plot(x_values, linspaceArray, "g")

axes2 = fig.add_axes([0.2,0.2,0.5,0.5])
axes2.plot(x_values, y_values_again, "r--")
```

```python
# most common usage
fig, axes = plt.subplots()
```

```python
colors = ["navy", "grey", "yellow", "lightgrey"] 
(fig, axes) = plt.subplots(2,2, figsize=(10,6), dpi=100)
for index, ax in enumerate(axes.flatten()):
    ax.plot(x_values, y_values, c=colors[index])
plt.tight_layout() # seperated - better visualizations
```

```python
colors = ["navy", "grey", "purple", "lightgrey"] 
(fig, axes) = plt.subplots(2,2, figsize=(10,6), dpi=100)
for index, ax in enumerate(axes.flatten()):
    if index % 2 == 0:
        ax.plot(x_values, y_values, c=colors[index], linestyle="-.")
    else:
        ax.plot(x_values, y_values, c=colors[index], linewidth=4)

plt.tight_layout()
```

```python
colors = ["navy", "grey", "purple", "lightgrey"] 
(fig, axes) = plt.subplots(2,2, figsize=(10,6), dpi=100)
for index, ax in enumerate(axes.flatten()):
    if index % 2 == 0:
        ax.plot(x_values, y_values, c=colors[index], linestyle="-.", markersize=20, marker="o")
    else:
        ax.plot(x_values, y_values, c=colors[index], linewidth=4,  markersize=20, marker="+")

plt.tight_layout()
```
